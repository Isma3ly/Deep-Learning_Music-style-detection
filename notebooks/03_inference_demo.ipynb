# 03 – Inference demo

This notebook demonstrates how to load the trained model and PCA, and predict the genre of a WAV file using `music_genre_classifier.inference`.

```python
import os
import logging
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"  # Expected on Windows: TF uses CPU
logging.getLogger("tensorflow").setLevel(logging.ERROR)

import sys
from pathlib import Path

# Ensure project root is on path so music_genre_classifier can be imported
_root = Path.cwd() if (Path.cwd() / "music_genre_classifier").is_dir() else Path.cwd().parent
if str(_root) not in sys.path:
    sys.path.insert(0, str(_root))

from music_genre_classifier.inference import predict_genre_from_file

MODELS_DIR = _root / "models"
weights_path = MODELS_DIR / "DNN3.weights.h5"
pca_path = MODELS_DIR / "pca"
# Paths under project root; GTZAN layout: data/genres_original/<genre>/*.wav
filler_track_path = str(_root / "data/genres_original/rock/rock.00000.wav")
```

```python
audio_path = str(_root / "data/genres_original/pop/pop.00000.wav")  # or any WAV under data/
genre, prob, conf_text = predict_genre_from_file(
    audio_path,
    model_weights=str(weights_path),
    pca_path=str(pca_path),
    filler_track_path=filler_track_path,
)
print(conf_text + genre, f"(p={prob:.2f})")
```
